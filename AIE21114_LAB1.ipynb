```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np


audio_file_path = r"C:\Users\Mohith Yashu\Downloads\mohith lab 1.wav"
signal, sr = librosa.load(audio_file_path)


plt.figure(figsize=(10, 5))
librosa.display.waveshow(signal, sr=sr)
plt.title('Speech Signal Waveform')
plt.show()
```

```python
print("Length of the signal:", len(signal), "samples")
print("Sampling rate:", sr, "Hz")
print("Magnitude range:", np.min(signal), "to", np.max(signal))
```

```python
import IPython.display as ipd

start_time = 1   
end_time = 5    

start_sample = int(start_time * sr)     
end_sample = int(end_time * sr)         
segment = signal[start_sample:end_sample]  

 
librosa.display.waveshow(segment, sr=sr)

 
ipd.Audio(segment, rate=sr)
```

```python
import IPython.display as ipd

segment_1_start_time = 1
segment_1_end_time = 3
segment_2_start_time = 2
segment_2_end_time = 3

segment_1_start = int(segment_1_start_time * sr)
segment_1_end = int(segment_1_end_time * sr)
segment_2_start = int(segment_2_start_time * sr)
segment_2_end = int(segment_2_end_time * sr)

segment_1 = signal[segment_1_start:segment_1_end]
segment_2 = signal[segment_2_start:segment_2_end]

print("Playing Segment 1:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_1, sr=sr)
plt.title('Segment-1')
plt.show()

ipd.Audio(segment_1, rate=sr)
```

```python
print("Playing Segment 2:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(segment_2, sr=sr)  
plt.title('Segment-2')  
plt.show()

ipd.Audio(segment_2, rate=sr)
```
